# Heuristic Search

```python
import heapq

goal = (
    (1, 2, 3),
    (8, 0, 4),
    (7, 6, 5)
)

start = (
    (2, 8, 3),
    (1, 6, 4),
    (7, 0, 5)
)

MOVES = [(-1, 0), (1, 0), (0, -1), (0, 1)]
def dispaly(puzzle):
    for row in puzzle:
      print(row)
    print()

def manhattan_distance(state):
    distance = 0
    for i in range(3):
        for j in range(3):
            tile = state[i][j]
            if tile != goal[i][j] and goal[i][j] != 0:
                distance += 1
    return distance

def get_neighbors(state):
    neighbors = []
    empty_pos = next((r, c) for r in range(3) for c in range(3) if state[r][c] == 0)
    r, c = empty_pos
    for dr, dc in MOVES:
        nr, nc = r + dr, c + dc
        if 0 <= nr < 3 and 0 <= nc < 3:
            new_state = [list(row) for row in state]
            new_state[r][c], new_state[nr][nc] = new_state[nr][nc], new_state[r][c]
            neighbors.append(tuple(tuple(row) for row in new_state))
    return neighbors

def heuristic_search(start, goal):
    # The open list (priority queue)
    open_list = []
    heapq.heappush(open_list, (manhattan_distance(start), start, []))  # (heuristic, state, path)

    # Set of visited states
    visited = set()
    visited.add(start)
    x = 0
    while open_list:
        x += 1
        _, current_state, path = heapq.heappop(open_list)

        # If goal state is reached, return the path
        if current_state == goal:
            # print(f"x = {x}")
            return path + [current_state]

        # Explore neighbors
        print("=============================================================")
        print("current state:")
        for step in current_state:
          print(step)
        print()

        print("neighbor:")
        for neighbor in get_neighbors(current_state):
            if neighbor not in visited:
                dist = manhattan_distance(neighbor)
                visited.add(neighbor)
                print(f"f(n) = h(n) + g(n) = {dist + x}")
                for step in neighbor:
                  print(step)
                print()
                heapq.heappush(open_list, (dist, neighbor, path + [current_state]))

    return None  # Return None if no solution is found

# Solve the 8 puzzle problem
solution = heuristic_search(start, goal)

# Print the solution path
if solution:
    print("=============================================================")
    print("Solution path is:")
    for step in solution:
        for row in step:
            print(row)
        print()  # Blank line between steps
else:
    print("No solution found.")
```

# Best First Search

```python
import heapq

class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v, cost):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []
        self.graph[u].append((v, cost))
        self.graph[v].append((u, cost))  # Assuming an undirected graph

    def best_first_search(self, start, goal, heuristic):
        pq = []
        heapq.heappush(pq, (heuristic[start], start))  # Priority queue (min-heap)
        visited = set()
        open_list = [(start, heuristic[start])]
        close_list = []

        print("| Step | Open List (h(n))       | C List   |")
        print("|------|------------------------|-------------|")
        step = 1

        while pq:
            _, current = heapq.heappop(pq)

            if current in visited:
                continue

            print(f"| {step:<4} | {', '.join(f'{node}[{h}]' for node, h in open_list) if open_list else '-':<22} | {', '.join(f'{node}[{h}]' for node, h in close_list) if close_list else '-'} |")
            step += 1

            visited.add(current)
            open_list = [(node, h) for node, h in open_list if node != current]
            close_list.append((current, heuristic[current]))

            if current == goal:
                print("\nSolution Path: " + " → ".join(f"{node}" for node, h in close_list))
                return

            for neighbor, _ in self.graph.get(current, []):
                if neighbor not in visited and neighbor not in [n for n, _ in open_list]:
                    heapq.heappush(pq, (heuristic[neighbor], neighbor))
                    open_list.append((neighbor, heuristic[neighbor]))

        print("Goal not reachable.")

# Example usage:
heuristic = {
    'S': 14, 'A': 12, 'B': 5, 'C': 7, 'D': 3, 'E': 8, 'F': 2, 'H': 4, 'I': 9, 'G': 0
}

g = Graph()
g.add_edge('S', 'A', 1)
g.add_edge('S', 'B', 1)
g.add_edge('A', 'C', 1)
g.add_edge('A', 'D', 1)
g.add_edge('B', 'E', 1)
g.add_edge('B', 'F', 1)
g.add_edge('E', 'H', 1)
g.add_edge('E', 'I', 1)
g.add_edge('F', 'G', 1)

g.best_first_search('S', 'G', heuristic)

```

# A* search

```python
import heapq

# Define the graph structure
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v, weight):
        if u not in self.graph:
            self.graph[u] = []
        self.graph[u].append((v, weight))

# Define the A* search algorithm
def a_star_search(graph, start, goal, heuristics):
    # Priority queue to store nodes to be explored, with (f(n), node)
    open_list = []
    heapq.heappush(open_list, (0 + heuristics[start], start))  # f(n) = g(n) + h(n), g(n) initially 0
    came_from = {}  # To store the path
    g_score = {start: 0}  # g(n), cost to reach start node is 0
    f_score = {start: heuristics[start]}  # f(n) = g(n) + h(n)
    close_list = set()  # Set to keep track of explored nodes


    while open_list:
        # Pop the node with the lowest f(n)
        _, current = heapq.heappop(open_list)

        # Add current node to closed list
        close_list.add(current)


        # If we reached the goal
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1]  # Return the path from start to goal

        # Explore neighbors
        if current in graph.graph:
            for neighbor, weight in graph.graph[current]:
                if neighbor in close_list:  # Skip nodes already in the closed list
                    continue

                tentative_g_score = g_score[current] + weight

                # If this path to the neighbor is better, update g and f scores
                if neighbor not in g_score or tentative_g_score < g_score[neighbor]:
                    came_from[neighbor] = current
                    g_score[neighbor] = tentative_g_score
                    f_score[neighbor] = tentative_g_score + heuristics[neighbor]
                    heapq.heappush(open_list, (f_score[neighbor], neighbor))

    return None  # If no path is found

# Create the graph from your input
g = Graph()
g.add_edge('S', 'A', 1)
g.add_edge('S', 'G', 10)
g.add_edge('A', 'B', 2)
g.add_edge('A', 'C', 1)
g.add_edge('B', 'D', 5)
g.add_edge('C', 'D', 3)
g.add_edge('C', 'G', 4)
g.add_edge('G', 'C', 4)

# Define the heuristics for each node
heuristics = {
    'S': 7,
    'A': 6,
    'B': 5,
    'C': 2,
    'D': 1,
    'G': 0
}

# Perform A* search from S to G
path = a_star_search(g, 'S', 'G', heuristics)

print("\nPath found:", end=" ")
for x in path:
    if x == path[-1]:
        print(x, end="")
    else:
        print(x, end=" -> ")
```
